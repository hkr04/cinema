```python
# Author: Zhang
import json
from openai import OpenAI
from datetime import datetime
from tqdm import tqdm

client = OpenAI(
    api_key="sk-xxxxx", # 替换为自己的 API 密钥
    base_url="https://api.deepseek.com",
)
```

```python
def get_data():
    system_prompt = """
    你需要为一个叫做“影响力”的智能影院助手构建电影问答的数据集。
    
    用户可能会以模糊的印象提问，如：“好像有桂纶镁出演，两女一男的青春电影。”
    
    模型需要回答最多 3 个相关的电影剧情简介，如：“为您找到相关的电影如下：\n\n**蓝色大门**\n\n《蓝色大门》是由易智言自编自导，桂纶镁、陈柏霖、梁又琳主演的剧情片。该片于2002年9月27日在台湾上映。该片讲述了高中生孟克柔、林月珍和张士豪之间懵懂的爱情以及成长的故事。”
    
    用户也可能会给出一些关键词让你帮忙推荐相关电影，如：“有一点恐怖和悬疑的元素，刺激一些的。”
    
    模型回复示例：“为您推荐**恐怖游轮**。\n\n剧情简介：单身母亲杰西（梅利莎·乔治　饰）留下孩子，同一帮朋友乘游艇出海游玩。出海后不久，游艇即不幸遇上风暴，当众人无计可施之时，一艘游轮经过，他们以为遇上救援，登上游艇，却发现这艘1930年失踪的神秘游轮让杰西陷入无尽轮回的磨难之中。”
    
    EXAMPLE JSON OUTPUT:
    {
        "instruction": "人类指令（必填）",
        "output": "模型回答（必填）",
    }
    """

    user_prompt = "不要局限于上述示例，尽情地发挥你的想象、尽可能地涉及多样的人类指令以及多样的电影，优先考虑日常化的人类指令和最新最热的电影。记住，人类指令大多较为模糊，因为人们自己也并不确定自己的需求。推荐电影时，一次可以推荐多个类似电影；介绍电影时，可以考虑加入各方的真实评价展开叙述。尽量用活泼生动的语言，确保该对话鲜活而真实。开始构建数据集！"

    messages = [{"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt}]

    response = client.chat.completions.create(
        model="deepseek-chat",
        messages=messages,
        temperature=1.5, # 如果使用 ChatGPT，将此项改为 0.95
        response_format={
            'type': 'json_object'
        }
    )

    return json.loads(response.choices[0].message.content)
```

```python
n_sample = 500
data_path = "./data/influence_9.json"

samples = []

for idx in tqdm(range(n_sample)):
    try:
        sample = get_data()
        samples.append(sample)
    except Exception as e:
        print(e)
        pass
    if idx % 10 == 0:
        with open(data_path, "w", encoding="utf-8") as f:
            json.dump(samples, f, indent=4)
        print(sample)
```

```python
samples = []
data_paths = ["data/influence_demo.json", "data/influence_1.json", "data/influence_2.json", "data/influence_3.json", "data/influence_4.json",
              "data/influence_5.json", "data/influence_6.json", "data/influence_7.json", "data/influence_8.json",
              "data/influence_9.json", "data/influence_10.json"]
for data_path in data_paths:
    with open(data_path, "r", encoding="utf-8") as f:
        samples.extend(json.load(f))
```

```python
for sample in samples:
    sample["input"] = ""
```
